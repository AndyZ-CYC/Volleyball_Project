### Estimation of Transition Matrix

```python
useful_columns = ['skill', 'evaluation_code', 'attack_code', 'serving_team', 'team', 'point']
```

```python
temp_df = df[useful_columns]
temp_df
```

```python
temp_df['side'] = temp_df.apply(lambda x: 'S' if x['serving_team'] == x['team'] else 'R' if pd.notna(x['team']) else None, axis=1)
temp_df
```

```python
temp_df.loc[temp_df['point'] == True, ['skill', 'evaluation_code']] = ['Point', '']
```

```python
temp_df['skill'].unique()
```

```python
skill_code_dict = {
    'Serve': 'SV',
    'Reception': 'R', 
    'Set': 'S', 
    'Attack': 'A', 
    'Dig': 'D', 
    'Point': 'P', 
    'Block': 'B', 
    'Freeball': 'F'
}
temp_df['skill_code'] = temp_df['skill'].map(skill_code_dict)
```

```python
temp_df.loc[temp_df['skill_code'].isna(), 'side'] = np.nan
```

```python
temp_df = temp_df[['skill_code', 'side', 'evaluation_code', 'attack_code']]
temp_df.head(20)
```

```python
def create_state_code(df):
    # Initialize the 'state' and 'prev_state' columns with NaN values
    df['state'] = np.nan
    df['prev_state'] = np.nan
    
    # Variable to store the state code of the previous row
    previous_state = None
    
    # Iterate over the dataframe row by row
    for i in range(len(df)):
        skill_code = df.at[i, 'skill_code']
        side = df.at[i, 'side']
        evaluation_code = df.at[i, 'evaluation_code']
        attack_code = df.at[i, 'attack_code']
        
        # If skill_code is NaN, this is the separator between points
        if pd.isna(skill_code):
            previous_state = None
            continue

        # Rule 2: If the skill_code is 'SV' (serve), start a new state code with 'S_SV'
        if skill_code == 'SV':
            df.at[i, 'prev_state'] = previous_state  # Store previous state
            df.at[i, 'state'] = 'S_SV'
            previous_state = 'S_SV'
            continue
        
        # Rule 3: If the skill_code is 'P' (serve), start a new state code with 'S_SV'
        if skill_code == 'P':
            df.at[i, 'prev_state'] = previous_state  # Store previous state
            df.at[i, 'state'] = f'{side}_P'
            previous_state = f'{side}_P'
            continue

        # Rule 4: If the side is the same as the previous row, append skill_code and evaluation_code
        if previous_state is not None and df.at[i, 'side'] == df.at[i-1, 'side']:
            if skill_code == 'A':  # Rule 5: Attack state
                df.at[i, 'prev_state'] = previous_state  # Store previous state
                df.at[i, 'state'] = f"{previous_state}.{skill_code}{attack_code}"
            else:
                df.at[i, 'prev_state'] = previous_state  # Store previous state
                df.at[i, 'state'] = f"{previous_state}.{skill_code}{evaluation_code}"
            previous_state = df.at[i, 'state']
            continue

        # Rule 5: If the side is different from the previous row, start a new state with 'f{side}_{skill_code}{evaluation_code}'
        if previous_state is None or df.at[i, 'side'] != df.at[i-1, 'side']:
            df.at[i, 'prev_state'] = previous_state  # Store previous state
            if skill_code == 'A':  # Rule 5: Attack state
                df.at[i, 'state'] = f"{side}_{skill_code}{attack_code}"
            else:
                df.at[i, 'state'] = f"{side}_{skill_code}{evaluation_code}"
            previous_state = df.at[i, 'state']

    return df
```

```python
def remove_consecutive_nan_rows(df):
    # Identify rows where all values are NaN
    nan_rows = df.isna().all(axis=1)

    # Create a mask to mark the start of each block of consecutive NaNs
    drop_mask = nan_rows & nan_rows.shift(fill_value=False)

    # Remove consecutive NaN rows but keep the first one
    df_cleaned = df[~drop_mask].copy()

    return df_cleaned

# Apply the function to clean the DataFrame
temp_df = remove_consecutive_nan_rows(temp_df)
temp_df = temp_df.reset_index(drop=True)
```
